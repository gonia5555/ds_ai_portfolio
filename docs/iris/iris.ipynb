##### The data includes measurements of four features: sepal length, sepal width, petal length, and petal width.

##### Each row in the dataset represents a single flower, with measurement values given in centimeters.

##### The dataset consists of 150 samples, with 50 samples for each species, and is widely used as a basic dataset for testing classification

* **sepal length** - Sepal length in cm
* **sepal width** - Sepal width in cm
* **petal length** - Petal length in cm
* **petal width** - Petal width in cm

##### Based on petal length and width, let’s create a simple function that, given the petal length and width, returns the Iris class:

- Setosa – very small petals (length < 2 cm, width < 0.6 cm)  
- Versicolor – medium petals (length ~3–5 cm, width < 1.8 cm)  
- Virginica – large petals (length > 5 cm, width < 1.4 cm)


```python
def classify_iris(petal_length, petal_width):
    if petal_length < 2.5:
        return "Iris-setosa"      # petal lenght < 2.5 cm → setosa 
    elif petal_width < 1.8:
        return "Iris-versicolor"   # If not Setosa, then consider the width < 1.8 cm → versicolor
    else:
        return "Iris-virginica"    # Other cases → virginica 

```

```python
# Checking accuracy using a loop with our classify_iris function, which iterates over all rows of the DataFrame.
correct = 0
total = len(df)

for i, row in df.iterrows():
    predicted = classify_iris(row['petal length'], row['petal width'])
    if predicted == row['class']:
        correct += 1

accuracy = correct / total * 100
print(f"Classification accuracy: {accuracy:.2f}%")
```

##### Considering the Iris petal dimensions (length and width) allows achieving about 96% accuracy in species identification, confirming that these are key features for effective Iris classification.
